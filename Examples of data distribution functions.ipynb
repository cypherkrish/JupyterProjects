# Generage some random numbers with a norman discribution

## "mu" is the desired mean
## "sigma" is the standarddeviation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu = 5.0
sigma = 2.0
```

```python
vlues = np.random.normal(mu, sigma, 10000)
```

```python
plt.hist(vlues, 50)
plt.show()
```

## Exponential PDF/ "Power Law"

```python
x
```

## Binomial Probability Mass Function

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 10, 0.001)
```

```python
x
```

```python
# Shape parameters for pmf(Probability Mass function)
n, p = 10, 0.5
```

```python
plt.plot(x, binom.pmf(x, n, p))
```

## Poission Probability Mass Function

```python
from scipy.stats import poisson
import matplotlib.pyplot as plt
import numpy as np
```

```python
mu = 500
x = np.arange(400, 600, 0.5)
```

```python
plt.plot(x, poisson.pmf(x, mu))
```
